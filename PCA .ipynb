# Principal Component Analysis (PCA):
## What is PCA?
* `PCA` is a statistical method that `transforms high-dimensional data` into a `new coordinate system`, where the new axes (principal components) are orthogonal and capture the maximum variance in the data.

# When to Use PCA:
## High-dimensional data: 
* When dealing with `datasets with a large number of features`, `PCA` can `help` in `reducing the dimensionality` while preserving most of the information.
## Data Visualization: 
* `PCA` can be used to `visualize high-dimensional` data in a `lower-dimensional space`.
## Noise Reduction:
* `PCA` can `reduce` the `noise` in the data by focusing on the `most significant features.`

# How Principal Component Analysis(PCA) works?

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()
X = iris.data

# Initialize PCA and fit-transform the data
pca = PCA(n_components=2,random_state=111)
X_pca = pca.fit_transform(X)
X_pca
```

```python
# Visualize the reduced data
import matplotlib.pyplot as plt
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.show()
```

## Advantages of PCA:
## Dimensionality Reduction: 
* PCA reduces the number of features while preserving the variance in the data.
## Computationally Efficient: 
* PCA is computationally efficient, making it suitable for large datasets.
