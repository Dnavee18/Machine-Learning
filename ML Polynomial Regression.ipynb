# ML Polynomial Regression

* Polynomial Regression is a `regression algorithm that models the relationship between a dependent(y) and independent variable(x) as nth degree polynomial`.
* The Polynomial Regression equation is given below:

* It is also called the ``special case of Multiple Linear Regression in ML``. Because we add some ``polynomial terms to the Multiple Linear regression equation to convert it into Polynomial Regression``.
* It is a ``linear model with some modification in order to increase the accuracy.``
* The dataset used in `Polynomial regression for training is of non-linear nature.`
* It makes use of a `linear regression model to fit the complicated and non-linear functions and datasets.`
* Hence, "In Polynomial regression, the ``original features are converted into Polynomial features of required degree (2,3,..,n) and then modeled using a linear model."``

## Need for Polynomial Regression:
The need of Polynomial Regression in ML can be understood in the below points:

* If we `apply a linear model on a linear dataset`, then it provides us a `good result as we have seen in Simple Linear Regression`, but if we apply the same model `without any modification on a non-linear datase`t, then it will produce a drastic output.Due to which `loss function will increase`, the `error rate will be high`, and `accuracy will be decreased`.
* So for such cases, `where data points are arranged in a non-linear fashion`, we need the `Polynomial Regression model.`
* We can understand it in a `better way using the below comparison diagram of the linear dataset and non-linear dataset.`


![](https://static.javatpoint.com/tutorial/machine-learning/images/machine-learning-polynomial-regression.png)

* In the above image, we have taken a dataset which is arranged non-linearly. So if we try to cover it with a linear model, then we can clearly see that it hardly covers any data point. On the other hand, a curve is suitable to cover most of the data points, which is of the Polynomial model.
* Hence, if the datasets are arranged in a non-linear fashion, then we should use the Polynomial Regression model instead of Simple Linear Regression.


## Steps for Polynomial Regression:
The main steps involved in Polynomial Regression are given below:

* Data Pre-processing
* Build a Linear Regression model and fit it to the dataset
* Build a Polynomial Regression model and fit it to the dataset
* Visualize the result for Linear Regression and Polynomial Regression model.
* Predicting the output.

```python
# importing libraries  
import numpy as nm  
import matplotlib.pyplot as mtp  
import pandas as pd  
  
#importing datasets  
data_set= pd.read_csv('Position_Salaries.csv')  
print(data_set)
#Extracting Independent and dependent Variable  
x= data_set.iloc[:, 1:2].values  
y= data_set.iloc[:, 2].values  
```

```python
print(x)
print(y)
```

```python
#Fitting the Linear Regression to the dataset  
from sklearn.linear_model import LinearRegression  
lin_regs= LinearRegression()  
lin_regs.fit(x,y)  
```

```python
x
```

```python
 #Fitting the Polynomial regression to the dataset  
from sklearn.preprocessing import PolynomialFeatures  
poly_regs= PolynomialFeatures(degree= 2)  
x_poly= poly_regs.fit_transform(x)  
lin_reg_2 =LinearRegression()  
lin_reg_2.fit(x_poly, y)  
```

```python
x_poly
```

```python
#Visulaizing the result for Linear Regression model  
mtp.scatter(x,y,color="blue")  
mtp.plot(x,lin_regs.predict(x), color="red")  
mtp.title("Bluff detection model(Linear Regression)")  
mtp.xlabel("Position Levels")  
mtp.ylabel("Salary")  
mtp.show()  
```

```python
#Visulaizing the result for Polynomial Regression  
mtp.scatter(x,y,color="blue")  
mtp.plot(x, lin_reg_2.predict(poly_regs.fit_transform(x)), color="red")  
mtp.title("Bluff detection model(Polynomial Regression)")  
mtp.xlabel("Position Levels")  
mtp.ylabel("Salary")  
mtp.show()
```

```python
 #Fitting the Polynomial regression to the dataset  
from sklearn.preprocessing import PolynomialFeatures  
poly_regs= PolynomialFeatures(degree= 3)  
x_poly= poly_regs.fit_transform(x)  
lin_reg_2 =LinearRegression()  
lin_reg_2.fit(x_poly, y)  
```

```python
#Visulaizing the result for Polynomial Regression  
mtp.scatter(x,y,color="blue")  
mtp.plot(x, lin_reg_2.predict(poly_regs.fit_transform(x)), color="red")  
mtp.title("Bluff detection model(Polynomial Regression)")  
mtp.xlabel("Position Levels")  
mtp.ylabel("Salary")  
mtp.show()
```

```python
 #Fitting the Polynomial regression to the dataset  
from sklearn.preprocessing import PolynomialFeatures  
poly_regs= PolynomialFeatures(degree= 5)  
x_poly= poly_regs.fit_transform(x)  
lin_reg_2 =LinearRegression()  
lin_reg_2.fit(x_poly, y)  
```

```python
#Visulaizing the result for Polynomial Regression  
mtp.scatter(x,y,color="blue")  
mtp.plot(x, lin_reg_2.predict(poly_regs.fit_transform(x)), color="red")  
mtp.title("Bluff detection model(Polynomial Regression)")  
mtp.xlabel("Position Levels")  
mtp.ylabel("Salary")  
mtp.show()
```

# Advantages of Polynomial Regression:

## Flexibility: 
* It can fit a w`ide range of curves to the data`.
## Accuracy: 
* It can provide accurate results when the `underlying relationship between variables is nonlinear.`

# Disadvantages of Polynomial Regression:

## Overfitting: 
* It's prone to `overfitting`, especially with `higher-degree polynomials`.
## Computational Complexity: 
* `Higher-degree polynomials can increase the computational complexity of the model.`


## Real-time Examples:

## Stock Market Analysis: 
* Polynomial regression can be used to analyze stock market trends where the relationship between time and stock prices might not be linear. For example, predicting future stock prices based on historical data.

## Climate Data Analysis: 
* Climate scientists often use polynomial regression to analyze temperature trends over time. The relationship between time and temperature may not be strictly linear due to various factors, and polynomial regression can capture these nuances.

## Economic Forecasting: 
* Economists use polynomial regression to model relationships between economic variables like GDP growth, inflation rates, and unemployment rates. These relationships are often non-linear and can be better captured using polynomial regression.

## Medicine and Healthcare: 
* Polynomial regression can be used in medical research to analyze the relationship between dosage and response in drug trials. It can also be applied in healthcare analytics to predict patient outcomes based on various health indicators.
